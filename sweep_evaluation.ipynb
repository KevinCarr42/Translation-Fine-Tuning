```python
import pandas as pd
```

```python
def display_sweep_performance(filename):
    param_columns = ['lr', 'lora_r', 'lora_alpha', 'lora_dropout', 'train_samples', 'eval_samples']
    
    df = pd.read_csv(filename)
    model = filename.split("_")[0]
    
    print(f"\n{model}\n")
    display(df.sort_values('final_eval_loss'))
    print(f"\noptimal hyperparameters for {model}:\n")
    print(
        df.sort_values('final_eval_loss').reset_index(drop=True).loc[0, param_columns].to_dict()
    )
```

```python
display_sweep_performance("m2m100_sweep.csv")
```
